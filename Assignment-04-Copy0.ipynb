## Q1:Discrete FFT and Power Spectrum

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

hourdata=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/AllHourly.txt')[[6,28],2:]

# remove data pairs with nan's
hourdata = np.transpose(hourdata)
bad_points = np.any(np.isnan(hourdata),axis=1)
hourdata = hourdata[~bad_points]
hourdata = np.transpose(hourdata)

dc = hourdata[0,:]
````

**1** Using the data from the hourly time series at Deep Cove (`dc`), show that the canned `fft` routine returns the correct discrete Fourier Transform of a suitable subset of the data (your choice on what subset).  You can do this graphically, by comparing the amplitude and the phase of the Fourier components, or simply show that the resulting numbers are the same.  

Hints: 
  - ignore bad data (rather than dealing with it "properly").  Make sure the units of your FT are: $\mathrm{^oC\, s}$.
  - If you want to compute the DFT effeciently, form the matrix $F$ referred to in the notes, and plot some of the time series associated with the different entries and make sure they look like Sines and Cosines and that the edge conditions are what you want


````python
N = 100
T = N*3600.
dt = 3600.

#subsample data
x = dc[:N]

f = np.arange(0.,N/2+1)/T

#dft from assignment 3
def dft(x,dt,f):
    t = dt*np.arange(0,len(x))
    T,F = np.meshgrid(t,f)
    
    # make the fourier matrices
    E = np.exp(-1j*2*np.pi*T*F)
    X = dt*np.dot(E,x)
    return X

# get the fourier components from the dft
X = dft(x,dt,f)

#get the fourier components from the canned fft
Xcan = np.fft.fft(x)*3600

#get the amplitude of the fourier components
ampX = np.abs(X)
ampXcan =np.abs(Xcan)

#get the phase of the fourier components
phaseX = np.arctan(np.imag(X)/np.real(X))
phaseXcan = np.arctan(np.imag(Xcan)/np.real(Xcan))

#comparison plots
fig = plt.figure(figsize=(6,12))
ax = fig.add_subplot(2,1,1)
ax.plot(ampX,label='dft')
ax.plot(ampXcan[:N/2+1],label='canned fft')
ax.set_xlabel('f [Hz]')
ax.set_ylabel('|X|')
ax.legend(loc=1)
ax.set_title('Amplitude comparison')

ax = fig.add_subplot(2,1,2)
ax.plot(phaseX,label='dft')
ax.plot(phaseXcan[:N/2+1],label='canned fft')
ax.set_xlabel('f [Hz]')
ax.set_ylabel(r'$\theta_{xy} [rad]$')
ax.legend(loc=1)
ax.set_title('Phase comparison')
````

The resulting plots are the same.

## Q2: Emperically derive distribution of spectral estimate

**1** Write a short function to create a "red" time series by taking the cumulative sum of Normally distributed random data, and plot one realization.  The number of data points $N$ can be relatively small (say 1000).

````python
def red(N):
    x = np.cumsum(np.random.randn(N))
    return x

x = red(1000)

fig, ax = plt.subplots(1,1)
ax.plot(x)
ax.set_xlabel('t')
ax.set_ylabel('x(t)')
````

**2** Write a second function to return the estimate of the power spectrum of $x$.  For posterity sake, include a value of the sampling frequency, and return the proper frequencies for the estimate.  Plot the resulting spectrum.

````python
def pow(x,fs):
    dt = 1./fs
    N = np.size(x)
    T = N*dt
    
    f = np.arange(N/2)/T
    X = dt*np.fft.fft(x)
    G = (2./T)*np.real(np.conj(X[:N/2])*X[:N/2])
    return G,f

fig,ax = plt.subplots(1,1)
ax.loglog(f,G)
ax.set_xlabel('f [Hz]')
ax.set_ylabel('$G_{xx}(f_k)$')
````

````python
Gxs = np.zeros((100,500))

fig,ax = plt.subplots(1,1)

for i in range(0,100):
    x = red(1000)
    Gxx, f = pow(x,fs)
    Gxs[i,:] = Gxx
    ax.loglog(f,Gxx,'.',color='black')

ax.set_xlabel('f [Hz]')
ax.set_ylabel('$G_{xx}(f_k)$')
````
