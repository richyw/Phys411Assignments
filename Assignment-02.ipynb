## Q1: Show that the correlation co-efficient is indeed related to the fraction of varaince explained by a linear dependence.

**1** To start, make a $N(0,1)$ random variable $x$, with 200,000 entries.  Then make a linearly dependent variable $y_i= a x_i + n_i$, where a is some amplitude, and $n_i$ is a random noise varaible.  Assume that $n_i$ is distributed as $N(0,n)$, where $n$ is the noise level.

Calculate the correlation co-efficient $r_{xy}$ between $x$ and $y$ for 40 different values of $n$, logarithmically spaced between 0.01 and 100. (hint use `np.logspace`).  

Plot $r_{xy}$ versus the inverse of the noise $n$ normalized by the standard devaition of the signal $ax_i$: $\frac{a s_x}{n}$   You may want to use `ax.loglog` to visualize the orders of magnitude.  

Comment on the resulting comparison.
<hr>

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

N = 200000
a = 1.
log_arr = np.logspace(-2,2,num=40)

x = np.random.randn(N)
x -= np.mean(x)

r_list = np.empty(0)
n_norm = np.empty(0)

for n in log_arr:

    ni = np.random.randn(N)*n
    y =a*x+ni
    y -= np.mean(y)
    sx = np.std(x,ddof=1)
    sy = np.std(y,ddof=1)
    r = np.sum(x*y)/(sx*sy*(N-1))
    r_list = np.append(r_list,r)
    n_norm = np.append(n_norm,a*sx/n)
  
    
fig = plt.figure(figsize=(6,6))
ax= fig.add_subplot(1,1,1)
ax.plot(r_list,n_norm,'.')
ax.set_xlabel('correlation coefficient $r_{xy}$')
ax.set_ylabel('$as_x/n$')
ax.loglog()
```

The correlation coefficient increases as the inverse of the noise increases (noise decreasing). As the noise approaches zero, the correlation coefficient approaches 1, which indicates a high amount of correlation between x & y

for larger values of a, the noise $n_i$ becomes less significant on comparison to the random variable $x_i$, because $x_i$ is multiplied by a magnitude of $a$. Therefore the correlation coefficient approaches 1 more quickly with decreasing noise.

Comment on the histigrams

```python
N = 10
b = np.empty(0)
s = np.empty(0)

for i in range(0,10001):
    x = np.random.randn(N)
    xbar = np.mean(x)
    x -= xbar
    n = 1.0
    a = 1.0

    y = a*x+np.random.randn(N)*n
    ybar = np.mean(y)
    y -= ybar

    bb = (np.sum(x*y)-N*xbar*ybar)/(np.sum(x**2)-n*xbar**2)
    b = np.append(b,bb)
    
    ss = np.sqrt(np.sum(y**2)/(N-2))
    s = np.append(s,ss)
    
    
    
bins = 50    
fig = plt.figure(figsize=(6,8))
ax = fig.add_subplot(2,1,1)
ax.hist(b,bins)
ax.set_title('slope distribution')
ax.set_xlabel('slope')
ax.set_ylabel('frequency')
ax = fig.add_subplot(2,1,2)
ax.hist(s,bins)
ax.set_title('slope variance')
ax.set_xlabel('variance')
ax.set_ylabel('frequency')
     
```
