### **Q1** Frequency-domain filtering

<div style='background:#F0F0F0'>  Consider a time series with a sharp jump in the middle as in the notes:
</div>

```python
import numpy.random as random
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
%matplotlib inline

N=1001
dt=1.
t=np.arange(0,N*dt,dt)

x = np.cumsum(random.randn(N)) # red noise
f0=0.03 # Hz
x = x + 40.*np.sin(f0*2*np.pi*t) # + sine wave
x=x-np.mean(x)
noise = 60.*random.randn(N)
xn = x+noise
xn[N/2:]=xn[N/2:]-2000.
# remove the mean!!
xn = xn-np.mean(xn)
```

<div style='background:#F0F0F0'>**1** Implement a low-pass filter of the data in the frequency domain, with a cutoff frequency of $f_c=0.06\ Hz$.</div>
   
  1. plot the signal and the smoothed signal (suggest zooming in on the step).
  2. plot the spectral response $|H(f)|$
  3. plot the phase response $\mathrm{angle}(H(f))$.

<div style='background:#F0F0F0'>
Be sure to comment on how good your filter is.
</div>

```python
X = np.fft.fft(xn)
fc = 0.06
ff = np.linspace(0,1./dt,N)
nfft = 128

#filter signals above cuttoff frequency
X[(ff>fc) & (ff<(1./dt-fc))]=1.e-15
y = np.real(np.fft.ifft(X))+np.mean(xn)
py,f = mlab.psd(y,Fs=1./dt,NFFT=nfft,noverlap=nfft/2)
px,f = mlab.psd(xn,Fs=1./dt,NFFT=nfft,noverlap=nfft/2)
Gxy,f = mlab.csd(xn,y,Fs=1./dt,NFFT=nfft,noverlap=nfft/2)
H = np.sqrt(py/px)
ang = np.angle(Gxy)

fig,ax = plt.subplots(3,1,figsize=(6,12))
ax[0].plot(t,xn,label='signal')
ax[0].plot(t,y,label='smoothed signal')
ax[0].set_xlim(450,550)
ax[0].set_xlabel('time [s]'); ax[0].set_ylabel('x(t)'); ax[0].legend()
ax[1].semilogx(f,H)
ax[1].set_xlabel('f [Hz]'); ax[1].set_ylabel('|H(f)|');
ax[2].semilogx(f,ang)
ax[2].set_xlabel('f [Hz]'); ax[2].set_ylabel('phase');

```

There is a quick roll off in frequency space, but the filtered signal has an overshoot at the step in the spatial domain. 

<div style='background:#F0F0F0'>**2** Now repeat the above but apply a Hanning window to the frequency window *before* you apply it to the FFT of $x_n$.  Comment on how well the filter is doing when you do this.  (Its helpful to keep the previous result up)
</div>

```python
#apply hanning window to frequency window
wind = np.zeros(N)
han = np.hanning(dt/fc*2)
for i in range(0,np.size(han)/2):
    wind[i]=han[i+np.size(han)/2]
    wind[N-np.size(han)/2+i]=han[i]

#plot window,zoom in on edges
fig,ax = plt.subplots(1,2,sharey=True)
fig.suptitle('Plot of window used')
ax[0].plot(wind)
ax[0].set_xlim(0,20)
ax[1].plot(wind)
ax[1].set_xlim(N-20,N)

X = np.fft.fft(xn)
#apply filter in frequency space

X = X*wind

y = np.real(np.fft.ifft(X))+np.mean(xn)

py,f = mlab.psd(y,Fs=1./dt,NFFT=nfft,noverlap=nfft/2)
px,f = mlab.psd(xn,Fs=1./dt,NFFT=nfft,noverlap=nfft/2)
Gxy,f = mlab.csd(xn,y,Fs=1./dt,NFFT=nfft,noverlap=nfft/2)
H = np.sqrt(py/px)
ang = np.angle(Gxy)

fig,ax = plt.subplots(3,1,figsize=(6,12))
ax[0].plot(t,xn,label='signal')
ax[0].plot(t,y,label='smoothed signal')
ax[0].set_xlim(450,550)
ax[0].set_xlabel('time [s]'); ax[0].set_ylabel('x(t)'); ax[0].legend()
ax[1].semilogx(f,H)
ax[1].set_xlabel('f [Hz]'); ax[1].set_ylabel('|H(f)|');
ax[2].semilogx(f,ang)
ax[2].set_xlabel('f [Hz]'); ax[2].set_ylabel('phase');



```

The filter drops off earlier and more slowly in frequency space. There is no overshoot/rippling of the filtered signal in the spatial domain, but the sharpness of the step and the low frequency information is not preserved.

<div style='background:#F0F0F0'>**3** So the Hanning window falls off a *lot* earlier and more slowly than a step function, so adjust the cutoff frequency to get a reasonable filter that preserves the amplitude of the sine wave at $f=0.03\ \mathrm{Hz}$.  Comment on the pros and cons of this approach.  
</div>

```python
fc = 0.008

#apply hanning window to frequency window
wind = np.zeros(N)
han = np.hanning(dt/fc*2)
for i in range(0,np.size(han)/2):
    wind[i]=han[i+np.size(han)/2]
    wind[N-np.size(han)/2+i]=han[i]

X = np.fft.fft(xn)

#apply filter in frequency space
X = X*wind

y = np.real(np.fft.ifft(X))+np.mean(xn)

py,f = mlab.psd(y,Fs=1./dt,NFFT=nfft,noverlap=nfft/2)
px,f = mlab.psd(xn,Fs=1./dt,NFFT=nfft,noverlap=nfft/2)
Gxy,f = mlab.csd(xn,y,Fs=1./dt,NFFT=nfft,noverlap=nfft/2)
H = np.sqrt(py/px)
ang = np.angle(Gxy)

fig,ax = plt.subplots(3,1,figsize=(6,12))
ax[0].plot(t,xn,label='signal')
ax[0].plot(t,y,label='smoothed signal')
ax[0].set_xlim(450,550)
ax[0].set_xlabel('time [s]'); ax[0].set_ylabel('x(t)'); ax[0].legend()
ax[1].semilogx(f,H)
ax[1].set_xlabel('f [Hz]'); ax[1].set_ylabel('|H(f)|');
ax[2].semilogx(f,ang)
ax[2].set_xlabel('f [Hz]'); ax[2].set_ylabel('phase');
```

The amplitude of the sine wave at 0.03Hz is preserved quite well, and the overshoot is still quite small.

## Q2: Implementing a bandpass filter

<div style='background:#F0F0F0'>**1** Create a red-noise process with 20000 data points, $dt=1\ s$.  Using a **truncated Dirichlet Kernel**, create a **bandpass** filter centered at $f_c=0.1\ Hz$ that has a passband that is approximately 0.1 Hz wide. Choose a filter length that is long enough to accomodate 0.1 Hz. 
</div>

   1. Plot the unshifted and shifted filters versus $k$, the time index.
   2. Plot the spectra, the response function $H^2(f)$ and the phase response
   3. Comment on your bandpass and if it indeed passed the right frequencies.

HINT: when you make the Dirichlet Kernel filter make sure that the $k=0$ case is treated "correctly" because $sin(0)=0$ and you will get a discontinuity.  Calculate what $h_0$ should be from [L'Hôpital's_rule](http://en.wikipedia.org/wiki/L'Hôpital's_rule).

```python
N = 20000
dt = 1.
fc = 0.1
nfft = N/100

#red noise
t = np.arange(0,N*dt,dt)
x = np.cumsum(random.randn(N))

k = np.array(range(-N/2,N/2))
m = k*1.
M = fc*2*N+1

i = np.arange(0,N,1)
h = np.zeros(N)
for ind in i:

    if (k[ind] == 0):
        h[ind] = M/N #calculated from lhopitals rule

    else:
        h[ind] = (1./N)*np.sin(np.pi*k[ind]*M/N)/np.sin(np.pi*k[ind]/N)

#shift filter to get bandpass       
mc = 0.1
hshift = 2*h*np.cos(2*mc*np.pi*k/N)        
 
#plot filters vs time index    
fig,ax=plt.subplots(1,1,figsize=(6,4))
ax.plot(k,h,label='unshifted filter') 
ax.plot(k,hshift,label='shifted filter')
ax.legend()
ax.set_xlabel('k')
ax.set_ylabel('h(k)')
ax.set_xlim(-100,100)

#trunicate filter
filtlen = 2000
h = h[np.abs(k)<filtlen+1]

#filter
y = np.convolve(xn,hshift,mode='same')[:-M/2]

pxx,f=mlab.psd(xn,Fs=1./dt,NFFT=nfft)
pyy,f=mlab.psd(y,Fs=1./dt,NFFT=nfft)
pxy,f=mlab.csd(xn,y,Fs=1./dt,NFFT=nfft)
H2 = pyy/pxx
angle = np.angle(pxy)/np.pi

fig,ax=plt.subplots(3,1,figsize=(6,12))
ax[0].loglog(f,pxx,label='raw')
ax[0].loglog(f,pyy,label='filtered')
ax[0].legend()
ax[0].set_ylabel('$G_{xx}$')
ax[0].set_xlabel('f [Hz]')
ax[1].loglog(f,H2)
ax[1].set_ylabel('$H^2 = G_{yy}/G_{xx}$')
ax[1].set_xlabel('f [Hz]')
ax[2].semilogx(f,angle)
ax[2].set_xlabel('f [Hz]')
ax[2].set_ylabel('phase / $\pi$')

```

my filter is not passing the correct frequencies. It appears to be a low-pass filter still.

HINT: Do a high pass and low pass.  Design using `ellipord` and `ellip`.  These functions allow you to try and design a bandpass directly, but I could not get it to work.  Do not be too draconian in your passband - i.e. make it relatively broad, but do try and suppress the peaks.
```python
dt = 60.
fNyq = 1./2./dt
nfft = 256
Rp = .1
Rs = 40.

#high pass
fph = 1/(24.5*3600.)
fsh = 1/(25.*3600.)

n,fn=signal.ellipord(fph/fNyq,fsh/fNyq,Rp,Rs)
b,a=signal.ellip(n,Rp,Rs,fn)
yh = signal.lfilter(b,a,dc)

#low pass
fpl = 1/(23.5*3600.)
fsl = 1/(23.*3600.)

n,fn=signal.ellipord(fpl/fNyq,fsl/fNyq,Rp,Rs)
b,a=signal.ellip(n,Rp,Rs,fn)
y = signal.lfilter(b,a,yh)

py,ff=mlab.psd(y-np.mean(y),NFFT=nfft,Fs=1./dt)
px,f=mlab.psd(dc-np.mean(dc),NFFT=nfft,Fs=1./dt)

fig,ax = plt.subplots(1,1)
ax.loglog(f,px,label='unfiltered')
ax.loglog(ff,py,label='filtered')
ax.legend()

```
