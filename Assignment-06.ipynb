### Error bounds on coherence estimates

<div style='background:#F0F0F0'>  The 95% confidence intervals for the coherence squared $\gamma^2_{xy}(f)$ between two variables $x(t)$ and $y(t)$ are approximately:
$$ \gamma^2_{xy} \pm \frac{\sqrt{2}\left(1+\gamma^2_{xy}\right)}{\left|\gamma_{xy} \right|\sqrt{n_d}}$$
where $n_d$ is the number of blocks used to calculate the coherence.  

You **may** use `matplotlib.mlab.csd` for the following question, but you had best use it correctly!
</div>

<div style='background:#F0F0F0'>**1** Make an input process $x(t)$ as random Gaussian noise, and the output process the sum of a noise source and a running average of the input: $y(t)=n(t)+\frac{1}{b}\int_{t-b}^t x(t')dt'$ and plot (Zoom so we can see whats going on, and label your plot).  Make sure you generate a large time series.</div> 

```python
N = 5096 #number of samples
b = 10 #length of running average
dt = 1. #time steps
noisescale = 0.2 #scale of noise

x = np.random.randn(N)
y = np.zeros(N)

#running average of input
for i in range(0,N):
    
    if i <= b:
        y[i] = np.sum(x[0:i+1])*dt/b
    
    else:
        y[i] = np.sum(x[i-b+1:i+1])*dt/b

# noise source        
y += np.random.randn(N)*noisescale
    
fig,ax = plt.subplots(1,1,figsize=(6,4))
ax.plot(y, label='x(t)')
ax.plot(x,label='y(t)')
ax.set_xlim(0,50)
ax.set_xlabel('time [s]')
ax.set_ylabel('f(x)')
ax.legend(loc='upper right')
```

<div style='background:#F0F0F0'>**2** Compute the coherence squared of the input and the output.  Make a plot with the spectra in the upper panel and the coherence squared in the middle panel and the phase in the bottom.

You will need to choose a value for your fft length.  Choose a relative low one to get a smooth plot.
</div>

```python
Nfft = 64 #fft length

Gff,f = mlab.psd(x,NFFT=Nfft,Fs=1.,window=mlab.window_hanning,noverlap=Nfft/2)
Gxx,f = mlab.psd(y,NFFT=Nfft,Fs=1.,window=mlab.window_hanning,noverlap=Nfft/2)
Gxf,f = mlab.csd(x,y,NFFT=Nfft,Fs=1.,window=mlab.window_hanning,noverlap=Nfft/2)

gamma2 = np.abs(Gxf)**2/(Gxx*Gff)
theta = np.arctan2(np.imag(Gxf),np.real(Gxf))

fig,ax=plt.subplots(3,1,sharex=True,figsize=(6,12))
ax[0].loglog(f,Gff,label='x(t)')
ax[0].loglog(f,Gxx,label='y(t)')
ax[0].set_ylabel('$G (f)$')
ax[0].legend(loc=3)
ax[1].semilogx(f,gamma2)
ax[1].set_ylabel('$\gamma^2(f)$')
ax[2].semilogx(f,theta/np.pi)
ax[2].set_xlabel('f [Hz]')
ax[2].set_ylabel('$\phi (f)/\pi$')
plt.tight_layout()
```

At low frequencies, the power spectra for the output function is lower than the input process. This is likely because low frequency information is lost due to the smoothing. The input and the output are coherent at low frequencies, and drops off as frequency increases. There are lows in the coherence squared for integer multiples of 0.1, which is 1/b. The phase is smooth where the signals are coherent, and jumps around where they are incoherent

<div style='background:#F0F0F0'>**3** Now repeat the above for three different values of NFFT, each 2 to 5 times as large as the next.  Plot the resulting coherence squared functions in separate panels.  Add the confidence intervals to your plot and comment on which peaks are significant.
</div>

```python
n = [64,256,512];
count = [0,1,2]

fig,ax=plt.subplots(3,1,sharex=True,figsize=(6,12))

for i in count:
    Gff,f = mlab.psd(x,NFFT=n[i],Fs=1.,window=mlab.window_hanning,noverlap=n[i]/2)
    Gxx,f = mlab.psd(y,NFFT=n[i],Fs=1.,window=mlab.window_hanning,noverlap=n[i]/2)
    Gxf,f = mlab.csd(x,y,NFFT=n[i],Fs=1.,window=mlab.window_hanning,noverlap=n[i]/2)
    gamma2 = np.abs(Gxf)**2/(Gxx*Gff)
    
    Nblock = 2*N/n[i]-1
    error = np.sqrt(2)*(1+gamma2)/(np.abs(np.sqrt(gamma2))*np.sqrt(Nblock))
    
    ax[i].semilogx(f,gamma2)
    ax[i].fill_between(f[1:],gamma2[1:]-error[1:],gamma2[1:]+error[1:],alpha=0.5)
    ax[i].set_ylim(-1,1)
    ax[i].set_ylabel('$\gamma^2(f)$')
    ax[i].set_title('NFFT = %i'%n[i])
    
ax[2].set_xlabel('f [Hz]')    
plt.tight_layout()
```

The confidence intervals are larger for higher values of NFFT, which makes sense, because less blocks are used in the calculation. For NFFT=64, the first three peaks appear significant. The other graphs do not appear to have significant peaks.

## Q2: Shear turbulence data

```python
time=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/time.txt')
ac=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/ac.txt')
shear=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/s.txt')
```

<div style='background:#F0F0F0'>Download the data abaove. `time` is a measurement time in seconds. `shear` is a little less than 300 s of data from a turbulence shear probe. Units are $s^{-1}$. `ac` is a suitably scaled acceleration of the vehicle containing the shear probes.

**1** Plot the shear time series and the accelerometer time series.  Comment on any similarities and differences.  Comment on the stationarity of the time series.</div>

```python
fig,ax = plt.subplots(2,1,figsize=(6,8),sharex=True)
ax[0].plot(time,ac,label='acceleration',zorder=2)
ax[0].set_ylabel('acc $[ms^{-2}]$')
ax[1].plot(time,shear,label='shear',zorder=1)
ax[1].set_ylabel('shear $[s^{-1}]$')
ax[1].set_xlabel('time [s]')
plt.tight_layout()
#ax.legend()
```

The shear data is wider when the acceleration is changing more rapidly. The shear data appears to be centered along the line shear=0, where the acceleration is almost always negative.

<div style='background:#F0F0F0'> **2** Compute the power spectra of the two signals and the coherence squared (you could do the phase too, but it is not too thrilling).  Comment on the spectral similarities and the coherence.</div>

```python
N = np.size(time)
dt = time[N-1]/N
n = 128

Gss,f = mlab.psd(shear,NFFT=n,Fs=1./dt,window=mlab.window_hanning,noverlap=n/2)
Gaa,f = mlab.psd(ac,NFFT=n,Fs=1./dt,window=mlab.window_hanning,noverlap=n/2)
Gas,f = mlab.csd(shear,ac,NFFT=n,Fs=1./dt,window=mlab.window_hanning,noverlap=n/2)
gamma2 = np.abs(Gas)**2/(Gaa*Gss)

fig,ax = plt.subplots(2,1,sharex=True,figsize=(6,8))
ax[0].loglog(f,Gss,label='shear')
ax[0].loglog(f,Gaa,label='acceleration')
ax[0].legend(loc='best')
ax[0].set_ylabel('$G (f)$')
ax[1].semilogx(f,gamma2)
ax[1].set_ylabel('$\gamma^2(f)$')
ax[1].set_xlabel('f [Hz]')
plt.tight_layout()

```

The shear spectrum has two main peaks, which it shares with the acceleration spectrum. The acceleration spectrum has three other smaller peaks, which do not appear to show up in the shear spectrum. The shear spectrum is smooth and constant at low frequencies, while the acceleration is smooth at very low frequencies, and drops gradualy. The spectrum is incohrent at these low frequencies (< 1Hz) and is coherent at the two shared peaks.

<div style='background:#F0F0F0'> **3** Compute a spectra that has the contamination from the vibration removed using the linear technique described in the notes. Comment on how well your correction worked.</div>

```python
Gnn = Gss - np.abs(Gas)**2/Gaa

fig,ax=plt.subplots(1,1,figsize=(6,4))
ax.loglog(f,Gnn,linewidth=2,label='corrected')
ax.loglog(f,Gaa,label='acceleration')
ax.loglog(f,Gss,label='shear')
ax.set_xlabel('f [Hz]')
ax.set_ylabel('G(f)')
ax.legend(loc='best')
```

The first peak in the shear data has been completely removed, The second (and much larger) peak has reduced in size, but is still visible. For the low frequency, incoherent part of the specturum, the corrected spectrum is idential to the shear spectrum.

<div style='background:#F0F0F0'>**4** Chose two separate periods of the time series by eye, one that is "turbulent" and one that is "quiet", and plot the raw and corrected shear spectra.  Comment on the differences and the effect of the correction on each spectral estimate.</div>

```python
#quiet slice
s = shear[50/dt:100/dt] 
a = ac[50/dt:100/dt]

Gss,f = mlab.psd(s,NFFT=n,Fs=1./dt,window=mlab.window_hanning,noverlap=n/2)
Gaa,f = mlab.psd(a,NFFT=n,Fs=1./dt,window=mlab.window_hanning,noverlap=n/2)
Gas,f = mlab.csd(s,a,NFFT=n,Fs=1./dt,window=mlab.window_hanning,noverlap=n/2)
Gnn = Gss - np.abs(Gas)**2/Gaa

fig,ax=plt.subplots(1,1,figsize=(6,4))
ax.loglog(f,Gnn,linewidth=2,label='corrected spectra')
ax.loglog(f,Gss,label='raw spectra')
ax.set_xlabel('f [Hz]')
ax.set_ylabel('G(f)')
ax.legend(loc='best')
ax.set_title('Quiet')

#turbulent slice
s = shear[225/dt:275/dt] 
a = ac[225/dt:275/dt]

Gss,f = mlab.psd(s,NFFT=n,Fs=1./dt,window=mlab.window_hanning,noverlap=n/2)
Gaa,f = mlab.psd(a,NFFT=n,Fs=1./dt,window=mlab.window_hanning,noverlap=n/2)
Gas,f = mlab.csd(s,a,NFFT=n,Fs=1./dt,window=mlab.window_hanning,noverlap=n/2)
Gnn = Gss - np.abs(Gas)**2/Gaa

fig,ax=plt.subplots(1,1,figsize=(6,4))
ax.loglog(f,Gnn,linewidth=2,label='corrected spectra')
ax.loglog(f,Gss,label='raw spectra')
ax.set_xlabel('f [Hz]')
ax.set_ylabel('G(f)')
ax.legend(loc='best')
ax.set_title('Turbulent')
```

For the quiet time period, the first peak in the raw spectrum is much more pronounced than in the turbulent time period. This might be because the vibrational noise overwhelms the turbulence measurements, for low turbulence. The correction in the first peak is more noticable for the quiet time series. The correction in second peak is more greater for the turbulent time series.
