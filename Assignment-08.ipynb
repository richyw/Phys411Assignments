### **Q1** Sampling theorem and aliasing

<div style='background:#F0F0F0'>**1** Create a band-limited white noise timeseries with $N=1000$ entries, $dt=0.1 s$, but with all frequencies greater than 0.5 Hz filtered.  Plot the power spectrum to make sure you got it correctly...
</div>

```python
N = 1000
dt = 0.1
fNyq = 1./(2.*dt)

t = np.arange(0,N*dt,dt)
x = random.randn(N)

B = 0.5

fp = B*0.7
fs = B*0.9

n,fn=signal.ellipord(fp/fNyq, fs/fNyq,.1, 60.)
b,a=signal.ellip(n,.1,60.,fn)
xBL = signal.filtfilt(b,a,x)

nfft=100

pxx,f=mlab.psd(x,NFFT=nfft,Fs=1./dt,noverlap=nfft/2.,window=mlab.window_hanning)
pxf,f=mlab.psd(xBL,NFFT=nfft,Fs=1./dt,noverlap=nfft/2.,window=mlab.window_hanning)

fig = plt.figure(figsize=(8,10))

ax = fig.add_subplot(2,1,2)
ax.semilogx(f,pxx,label='raw',color='k',alpha=0.5)
ax.semilogx(f,pxf,label='bandlimited')
plt.axvline(x=0.5,linestyle='--',color='k')
ax.set_xlabel('f [Hz]');ax.set_ylabel('$G_{xx}$');ax.set_title('Power Spectrum')
ax.legend(loc=4)


ax = fig.add_subplot(2,1,1)
ax.plot(t,x,label='raw',color='k',alpha=0.5)
ax.plot(t,xBL,label='bandlimited')
ax.set_xlabel('t [s]'); ax.set_title('Time Series')
ax.legend(loc=4)
plt.tight_layout()
```

## Q2: Cubic Spline

Using the data below...

```python
t = np.array([0.,2.5,2.7,3.4,4.2,6.5,9.0])
x = np.cos(t*2*np.pi/3)
```

<div style='background:#F0F0F0'>**1** Perform a cubic spline interpolation onto an even grid from 0 to 9.0 by 
  1. creating the matrix from the notes
  2. inverting and solving for the cubic co-efficients
  3. using those polynomials to evaluate the interpolated data.
Compare with using `scipy.interpolate.interp1d`.
Comment on any similarities or differences between the two results.

```python
N = np.size(x)

delta = np.zeros(N-1)

for i in range(0,N-1):
    delta[i] = t[i+1]-t[i]

#create matrices from notes

A = np.zeros((N,N)) 
A[0,0] = 1
A[N-1,N-1] = 1

for i in range(1,N-1):
    A[i,i-1] = delta[i-1]
    A[i,i] = 2*(delta[i-1]+delta[i])
    A[i,i+1] = delta[i]

y = np.zeros(N)

for i in range(1,N-1):
    y[i] = 3.*((x[i+1]-x[i])/delta[i]-(x[i]-x[i-1])/delta[i-1])

#invert matrix and solve for cubic coefficients
c = np.dot(np.linalg.inv(A),y)

#print A
#print b
#print c

a = np.zeros(N)
b = 0.*a
d = 0.*a

#solve for other coefficients

for i in range(0,N-1):
    a[i] = x[i]
    d[i] = (c[i+1]-c[i])/3.
    b[i] = (x[i+1]-x[i]-delta[i]**2*c[i]-delta[i]**3*d[i])/delta[i]

#evaluate polynomial at evenly spaced time    
tn = np.arange(0.,9.1,0.1) 
NN = np.size(tn)

q = 0.*tn

for i in range(0,NN):
    if tn[i] < t[1]:
        q[i] = a[0] + b[0]*(tn[i]-t[0]) + c[0]*(tn[i]-t[0])**2 + d[0]*(tn[i]-t[0])**3
    elif tn[i] < t[2]:
        q[i] = a[1] + b[1]*(tn[i]-t[1]) + c[1]*(tn[i]-t[1])**2 + d[1]*(tn[i]-t[1])**3
    elif tn[i] < t[3]:
        q[i] = a[2] + b[2]*(tn[i]-t[2]) + c[2]*(tn[i]-t[2])**2 + d[2]*(tn[i]-t[2])**3
    elif tn[i] < t[4]:
        q[i] = a[3] + b[3]*(tn[i]-t[3]) + c[3]*(tn[i]-t[3])**2 + d[3]*(tn[i]-t[3])**3
    elif tn[i] < t[5]:
        q[i] = a[4] + b[4]*(tn[i]-t[4]) + c[4]*(tn[i]-t[4])**2 + d[4]*(tn[i]-t[4])**3
    elif q[i] < t[6]:
        q[i] = a[5] + b[5]*(tn[i]-t[5]) + c[5]*(tn[i]-t[5])**2 + d[5]*(tn[i]-t[5])**3

#canned routine
f = interpolate.interp1d(t,x,kind='cubic')
xcan = f(tn)

#comparison plot
fig,ax = plt.subplots(1,1,figsize=(8,6))
ax.plot(tn,q,label='manual cubic spline')
ax.plot(tn,xcan,color='red',label='canned cubic spline')
ax.plot(t,x,'.',color='k',marker='o',label='original data points')
ax.plot(tn,np.cos(tn*2*np.pi/3),'k--',label='true function')
ax.set_xlabel('t'); ax.set_ylabel('x(t)'); ax.set_title('Interpolation Comparison')
ax.legend(loc=0)        
        
```

The manual cubic spline interpolation appears to be very different than the canned routine. The canned fit appears more smooth. The manual fit looks like it might have sharp points at the original data points, which would suggest that the second derivative of the polynomial is not continious (a requirement of the cubic spline), but it could just be chaning quickly. 

The manual cubic spline is closer to the cosine function in many of the regions of the graph

Where the time gaps are smaller (ie, between 2.5 and 4.2 seconds), the canned routine and the manual fit are very similar.
