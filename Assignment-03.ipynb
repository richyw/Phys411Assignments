## Q1: Lag correlations of weather time series

````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
%matplotlib inline

hourdata=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/AllHourly.txt')[[6,28],2:]

# remove data pairs with nan's
hourdata = np.transpose(hourdata)
bad_points = np.any(np.isnan(hourdata),axis=1)
hourdata = hourdata[~bad_points]
hourdata = np.transpose(hourdata)

dc = hourdata[0,:]
jb = hourdata[1,:]
````

**1** From the hourly time series at Deep Cove (`dc`), calculate the *lag correlation co-efficient* $\rho_{xy}(\tau)$ out to a lag of $\tau=30\ \mathrm{days}$ (Umm, remember this data is hourly!).  (also rememebr to remove the mean of $dc$ before trying to to the calculations).  

Comment on the resulting plot.

````python
dc -= np.mean(dc)
lags = np.arange(30*24)

cdd=0.*lags

for ind,tau in enumerate(lags):
    if tau==0:
        cdd[ind]=np.mean(dc*dc)/np.var(dc)
    else:    
        cdd[ind] = np.mean(dc[:-tau]*dc[tau:])/np.var(dc)
    
fig,ax=plt.subplots(1,1)
ax.plot(lags,cdd)
ax.set_xlabel(r'$\tau [hours]$')
ax.set_ylabel('$p_{dd} [V^2]$')    



````

The lag correlation coefficient is periodic, with a period of one day. In addition to this daily trend, the correlation coefficient decreases overall as time increases. This suggests that the temperature at the same time each day is more correlated than the temperatures at different times of the day. It also suggests that the temperatures from close dates are more correlated than the temperatures from days spaced far apart. The correlation coefficient is unity at $\tau = 0$, as expected for autocorrilation.

*2* Now, instead of doing the whole year, look at just the winter months and the summer months and compare the lag correlations (no need to do this for the whole time series, just choose one winter and one summer).  Make sure to indicate which lag correlation is which on the plots.

Comment below on the different charcter of these two lag correlations.

````python
#winter (Dec, Jan, Feb)
#summer (June,July,Aug)
dc = hourdata[0,:]

dcw = dc[334*24:(334+88)*24]
dcw -= np.mean(dcw)
dcs = dc[151*24:(151+91)*24]
dcs -= np.mean(dcs)

cww = 0.*lags
css = 0.*lags

for ind,tau in enumerate(lags):
    if tau==0:
        cww[ind]=np.mean(dcw*dcw)/np.var(dcw)
        css[ind]=np.mean(dcs*dcs)/np.var(dcs)
    else:    
        cww[ind] = np.mean(dcw[:-tau]*dcw[tau:])/np.var(dcw)
        css[ind] = np.mean(dcs[:-tau]*dcs[tau:])/np.var(dcs)
               
fig = plt.figure(figsize=(6,12))
ax=fig.add_subplot(2,1,1)
ax.plot(lags,cww)
ax.set_xlabel(r'$t [hours]$')
ax.set_ylabel('$ p_{ww} [V^2]$')
ax.set_title('Winter months')
ax=fig.add_subplot(2,1,2)
ax.plot(lags,css)
ax.set_xlabel(r'$t [hours]$')
ax.set_ylabel('$ p_{ss} [V^2]$')
ax.set_title('Summer months')
````

The graphs still show the same daily periodicity as the yearly data. Overall, the corellation coefficient approaches zero more rapidly than in the first graph. This might be because only one year was considered in these cases.
